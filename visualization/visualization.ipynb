Bar chart

```python
labels = ["a","b","c"]
values =[1,4,3]
plt.bar(labels,values, color="k")
plt.show()
```

```python
lables = ["A", "B", "C"]
values = [1,4 ,2]
bars = plt.bar(lables, values, color = "#034dad")
bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")
plt.show()
```

```python
lables = ["A", "B", "C"]
values = [1,4 ,2]
bars = plt.bar(lables, values)
bars[0].set_hatch('o')
plt.show()
```

```python
lables = ["A", "B", "C"]
values = [1,4 ,2]
bars = plt.bar(lables, values)
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```

```python
lables = ["A", "B", "C"]
values = [1,4 ,2]
bars = plt.bar(lables, values)
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,60]
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
plt.show()
```

pie chart

```python
exp_vals = [1400,600,300,300,250] #expenses 
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.2f%%',radius=1.6, )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.4f%%',radius=1.5,explode=[.2,.2,.17,.15,0.19])
plt.show()
```
